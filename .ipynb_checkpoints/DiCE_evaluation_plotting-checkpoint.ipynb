```python
import matplotlib.pyplot as pl
import seaborn as sns
import os
import pandas as pd
import pickle
import itertools
```

# Figure 1 nonlinear plot

```python
figure1_results = {}
datasets = ['adult', 'german', 'compass', 'lending']
desired_classes = ['class1_g50k', 'class1_bad', 'class0_wont_recidivate', 'class1_Paid'] # for german, I swapped the two classes in the code

for data, d_class in zip(datasets, desired_classes):
    with open(os.path.join('figure1_summary_stats/' + data + '_nonlinear_' + d_class  +'.data'), 'rb') as filehandle:
        temp = pickle.load(filehandle)
    figure1_results[data] = temp
```

```python
eval_metrics = ['valid_unique_CFs', 
                'cate_div_count', 'cont_div_mad', 'cont_div_count',
                'cate_prox_count', 'cont_prox_mad', 'cont_prox_count'] 
```

```python
# Figure 1 plot  
x = [1,2,4,6,8,10]
dataset = ['adult', 'lending', 'german', 'compass']
fig, axes = pl.subplots(len(dataset), len(eval_metrics), 
                       figsize=(4*(len(eval_metrics)),4*(len(dataset))),
                       sharex=True, sharey=False)

axes = axes.flatten()
pl.xticks([1,2,4,6,8,10])

pl.rcParams.update({'axes.labelsize': 15})


xidx = [i for i in range(21,28)]
yidx = [0,7,14,21]

xlabel_dict = dict(zip(xidx, ['% Valid CFs', 'Cat-Count-Diversity', 'Continuous-Diversity', 'Cont-Count-Diversity',
                              'Cat-Prox-Sparsity', 'Continuous-Proximity', 'Cont-Prox-Sparsity']))
ylabel_dict = dict(zip(yidx, ['Adult-Income', 'LendingClub', 'German-Credit', 'COMPAS']))
        
    
axes_idxs = 0
for data, metric in itertools.product(dataset,eval_metrics):

    y1 = figure1_results[data]['DiverseCF']['without_postfix'][metric]
    y2 = figure1_results[data]['DiverseCF']['with_postfix'][metric]
    
    
    y3 = figure1_results[data]['NoDiverseCF']['without_postfix'][metric]
    y4 = figure1_results[data]['NoDiverseCF']['with_postfix'][metric]
    y5 = figure1_results[data]['RandomInitCF']['without_postfix'][metric]
    y6 = figure1_results[data]['RandomInitCF']['with_postfix'][metric]
    y7 = figure1_results[data]['RandomInitCF']['with_postfix'][metric][0]
    y7 = [y7]
    
    if metric in ['cate_prox_count', 'cont_prox_count']:
        y1 = [1-i for i in y1]
        y2 = [1-i for i in y2]
        y3 = [1-i for i in y3]
        y4 = [1-i for i in y4]
        y5 = [1-i for i in y5]
        y6 = [1-i for i in y6]
        y7 = [1-i for i in y7]
        
    if metric == 'cont_prox_mad':
        y1 = [-i for i in y1]
        y2 = [-i for i in y2]
        y3 = [-i for i in y3]
        y4 = [-i for i in y4]
        y5 = [-i for i in y5]
        y6 = [-i for i in y6]
        y7 = [-i for i in y7]

    if data == 'compass':
        y3 = [i if i!=0 else np.nan for i in y3]
        y4 = [i if i!=0 else np.nan for i in y4]
        y5 = [i if i!=0 else np.nan for i in y5]
        y6 = [i if i!=0 else np.nan for i in y6]
            
        
    a = axes[axes_idxs].plot(x, y1, '-D', alpha=1.0, label='DiverseCF', markersize=10, markerfacecolor='none')
    b = axes[axes_idxs].plot(x, y2, '--D', alpha=1.0,  label='DiverseCF-Sparse', markersize=10, markerfacecolor='none', color=a[0].get_color())
    c = axes[axes_idxs].plot(x, y3, '-o', alpha=1.0,  label='NoDiverseCF', markersize=10, markerfacecolor='none')
    d = axes[axes_idxs].plot(x, y4, '--o', alpha=1.0,  label='NoDiverseCF-Sparse', markersize=10, markerfacecolor='none', color=c[0].get_color())
    e = axes[axes_idxs].plot(x, y5, '-x', alpha=1.0,  label='RandomInitCF', markersize=10, markerfacecolor='none')
    f = axes[axes_idxs].plot(x, y6, '--x', alpha=1.0,  label='RandomInitCF-Sparse', markersize=10, markerfacecolor='none', color=e[0].get_color())
    g = axes[axes_idxs].plot(x[0], y7, '-s', alpha=1.0,  label='SingleCF', markersize=15, markerfacecolor='none', mew=2)
    
    if axes_idxs in xlabel_dict:
        axes[axes_idxs].text(0.5,-0.25, '#CFs', size=20, ha="center", 
                    transform=axes[axes_idxs].transAxes)
        axes[axes_idxs].text(0.5, 4.7, xlabel_dict[axes_idxs], size=20, ha="center", 
                    transform=axes[axes_idxs].transAxes) # 0.32,3.3
    if axes_idxs in ylabel_dict:
        axes[axes_idxs].set_ylabel(ylabel_dict[axes_idxs], fontsize=20)
        axes[axes_idxs].yaxis.set_label_coords(-0.2, 0.5)
        
    #axes[axes_idxs].set_ylim(ymin=0, ymax=7)
    axes[axes_idxs].tick_params(labelsize=12)
    
    if axes_idxs%7 == 0:
        axes[axes_idxs].set_ylim(ymin=0, ymax=110)
    elif axes_idxs%7 == 1:
        axes[axes_idxs].set_ylim(ymin=0, ymax=1.1)
    elif axes_idxs%7 == 2:
        axes[axes_idxs].set_ylim(ymin=0, ymax=6)
        #axes[axes_idxs].set_yticks(np.arange(0,1,0.1))
    elif axes_idxs%7 == 3:
        axes[axes_idxs].set_ylim(ymin=0, ymax=1.1) #ymin=-4.5, ymax=0
        #axes[axes_idxs].set_yticks(np.arange(0,1,0.1))
    elif axes_idxs%7 == 4:
        axes[axes_idxs].set_ylim(ymin=0, ymax=1.1)
    elif axes_idxs%7 == 5:
        axes[axes_idxs].set_ylim(ymin=-5, ymax=0) #ymin=-4.5, ymax=0
        #axes[axes_idxs].set_yticks(np.arange(0,1,0.1))
    elif axes_idxs%7 == 6:
        axes[axes_idxs].set_ylim(ymin=0, ymax=1.1) #ymin=-4.5, ymax=0
        #axes[axes_idxs].set_yticks(np.arange(0,1,0.1))
        
    axes_idxs += 1


handles, labels = axes[0].get_legend_handles_labels() #0.99
fig.legend(handles, labels, loc='upper center', ncol=4, bbox_to_anchor=(0.4, 0.08), 
           fancybox=True, shadow=True, fontsize=20)

#fig.tight_layout()
fig.subplots_adjust(top=0.93)
fig.subplots_adjust(bottom=0.15)
```

```python
outdir = 'paper_plots/'
filename = 'figure1_nonlinear_bad_to_good.pdf'
if not os.path.exists(outdir):
    os.makedirs(outdir)

path = os.path.join(outdir, filename)
fig.savefig(path, dpi=1000, bbox_inches='tight')
```

# Figure 2 nonlinear plot

```python
with open(os.path.join('figure2_summary_stats/figure2_summary_without_True_user_input_100.data'), 'rb') as filehandle:
    figure2_results = pickle.load(filehandle)
```

```python
model_type = 'nonlinear'
postfix = 'without'
lime_discretize = 'True'
cont_radius = 'user_input' 
discr_param = 1.0

postfix_param = postfix + '_postfix'
lime_discretize_param = 'discretize_' + lime_discretize

lr_dict = {'adult_linear': 0.05, 'adult_nonlinear': 0.05, 'german_linear': 0.05, 'german_nonlinear': 0.05, 
           'compass_linear': 0.05, 'compass_nonlinear': 0.05, 'lending_linear':0.05, 'lending_nonlinear': 0.05}
    
cont_radius_dict = {'adult': ['[5, 2]', '[10, 4]', '[20, 8]'], 
                    'german': ['[3, 544, 1, 1, 4]', '[6, 1088, 2, 2, 7]', '[12, 2176, 3, 3, 14]'], 
                    'compass': ['[1]', '[2]', '[4]'], 
                    'lending': ['[1, 9500, 1, 2]', '[3, 19000, 3, 4]', '[6, 38000, 6, 8]']}


datasets = ['adult', 'lending', 'german', 'compass']
algorithms = ['DiverseCF', 'NoDiverseCF', 'RandomInitCF', 'LIME']
all_total_CFs = [1,2,4,6,8,10]
```

```python
# for german, I swapped the two classes in the code
data_dclass_mapping = dict(zip(['adult', 'german', 'compass', 'lending'], 
                               ['class1_g50k', 'class1_bad', 'class0_wont_recidivate', 'class1_Paid'])) 
transit = 'bad_to_good'
```

## all mads

```python
#figure 2 all mads

markers = dict(zip(algorithms, ['D', 's', 'x', 'o']))
figure2_figs = []

for metric in ['precision', 'recall', 'f1']:

    x = all_total_CFs
    mad_len = 3
    xidx = [9,10,11]
    yidx = [0,3,6,9]
    xlabel_dict = dict(zip(xidx, ['0.5 MAD', '1 MAD', '2 MAD']))
    font1 = 20
    font2 = 15
    bbox = [0.46, 0.08]
    
    fig, axes = pl.subplots(len(datasets), len(cont_radius_dict[datasets[0]][0:mad_len]),
                   figsize=(4*(len(cont_radius_dict[datasets[0]][0:mad_len])), 4*(len(datasets))),
                   sharex=True, sharey=False)

    axes = axes.flatten()
    pl.xticks(all_total_CFs)

    pl.rcParams.update({'axes.labelsize': 15})
        
    ylabel_dict = dict(zip(yidx,['Adult-Income', 'LendingClub', 'German-Credit', 'COMPAS']))

    axes_idxs = 0
    for data in datasets:
        for cont_radius in cont_radius_dict[data][0:mad_len]:

            for algorithm in algorithms:

                if metric != 'accuracy':
                    key = data + algorithm + cont_radius + str(int(discr_param*100)) +  'CF_' + metric
                    if algorithm == 'DiverseCF':
                        y1 = figure2_results[data_dclass_mapping[data]][key]
                    else:
                        y1 = figure2_results[data_dclass_mapping[data]][key]
                    if algorithm == 'NoDiverseCF':
                        y1 = y1[0:6]

                    a = axes[axes_idxs].plot(x, y1, '-' + markers[algorithm], alpha=1.0, label=algorithm + ': CF_class' , markersize=10, markerfacecolor='none')
                    
                else:
                    key = data + algorithm + cont_radius + str(int(discr_param*100)) +  metric
                    if algorithm == 'DiverseCF':
                        y1 = figure2_results[data_dclass_mapping[data]][key]
                    else:
                        y1 = figure2_results[data_dclass_mapping[data]][key]
                    if algorithm == 'NoDiverseCF':
                        y1 = y1[0:6]

                    a = axes[axes_idxs].plot(x, y1, '-' + markers[algorithm], alpha=1.0, label=algorithm, markersize=10, markerfacecolor='none')

            if axes_idxs in xlabel_dict:
                axes[axes_idxs].text(0.5,-0.25, '#CFs', size=15, ha="center",
                            transform=axes[axes_idxs].transAxes)
                axes[axes_idxs].text(0.5, 4.7, xlabel_dict[axes_idxs], size=15, ha="center",
                            transform=axes[axes_idxs].transAxes)
            if axes_idxs in ylabel_dict:
                axes[axes_idxs].set_ylabel(ylabel_dict[axes_idxs], fontsize=font1)
                axes[axes_idxs].yaxis.set_label_coords(-0.2, 0.5)

            if axes_idxs%3 == 0:
                axes[axes_idxs].set_ylim(ymin=0, ymax=1.1)
            elif axes_idxs%3 == 1:
                axes[axes_idxs].set_ylim(ymin=0, ymax=1.1)
            elif axes_idxs%3 == 2:
                axes[axes_idxs].set_ylim(ymin=0, ymax=1.1)

            axes_idxs += 1


    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper center', ncol=2, bbox_to_anchor=(bbox[0], bbox[1]),
       fancybox=True, shadow=True, fontsize=font2)
    fig.subplots_adjust(top=0.93)
    fig.subplots_adjust(bottom=0.15)
    
    figure2_figs.append(fig)
```

```python
for figs, metric in zip(figure2_figs, ['precision', 'recall', 'f1']):
    outdir = 'paper_plots/all_mads'
    filename = 'figure2_'+ metric + transit +'.pdf'
    if not os.path.exists(outdir):
        os.makedirs(outdir)

    path = os.path.join(outdir, filename)
    figs.savefig(path, dpi=1000, bbox_inches='tight')
```
